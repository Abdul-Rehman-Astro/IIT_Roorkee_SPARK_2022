Importing all needed libraries

```python
import numpy as np
from PIL import Image
import os
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
```

```python
from matplotlib.pyplot import figure
from PIL import Image


input_image = Image.open("input.tif") #load an image file
input = np.array(input_image)
print(input.size)
print(input.shape)
figure(figsize=(20, 15), dpi=100)
plt.imshow(input)
plt.show()
```

```python
input_gray=cv2.cvtColor(input,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Scaled Image",input_gray)
# plt.imshow(input_gray)
# plt.show()
```
