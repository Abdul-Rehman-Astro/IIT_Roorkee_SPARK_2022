```python
import numpy as np
from PIL import Image
import os
import cv2
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
```

```python
from matplotlib.pyplot import figure
from PIL import Image


input_image = Image.open("input.tif") #load an image file
input = np.array(input_image)
figure(figsize=(10,10), dpi=100)
plt.imshow(input)
plt.show()
```

Defining Kernel for Morphology operations

```python
# Let's define our kernel size
kernel = np.ones((3,3), np.uint8)
```

# Erosion


```python
erosion = cv2.erode(input, kernel, iterations = 1)

# plt.subplot(3, 2, 2)
figure(figsize = (10,10),dpi=100)
plt.title("Erosion")
plt.imshow(erosion)
```

# Diltation

```python
dilation = cv2.dilate(input, kernel, iterations = 1)
# plt.subplot(3, 2, 3)
figure(figsize=(10,10),dpi=100),
plt.title("Dilation")
plt.imshow(dilation)
```

```python
# Let's define our kernel size
# kernel = np.ones((5,5), np.uint8)
```

# Opening (for removing noise)

```python
# Opening - Good for removing noise
opening = cv2.morphologyEx(input, cv2.MORPH_OPEN, kernel)
# plt.subplot(3, 2, 4)
figure(figsize = (10,10),dpi=100)
plt.title("Opening")
plt.imshow(opening)
```

# Closing (removing noise)

```python
# Closing - Good for removing noise
closing = cv2.morphologyEx(input, cv2.MORPH_CLOSE, kernel)
# plt.subplot(3, 2, 5)
figure(figsize=(10,10),dpi=100)
plt.title("Closing")
plt.imshow(closing)
```
